```python
import numpy as np
import symengine as sm
from math import sqrt, log

import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import cm

from pandas.plotting import table 
```

```python
def Lagrangian_print(f, g, s, h, param_b, sigma, lamda):
    L = f
    ln = 0
    for i in range(len(g)):
        L += sigma[i]*(g[i]+s[i])
        ln += sm.sympify(f'log({s[i]})')
    for i in range(len(h)):
        L += lamda[i]*h[i] 
    L -= param_b*ln
    fln = f - param_b*ln
    return L, fln
```

```python
def score_matrix(f, x, y):
    a = []
    for i in range(len(y)):
        a.append((x[i], y[i]))
    if type(f) == list:
        g = []
        for j in f:
            g.append(j.subs(dict(a)))
    else:
        g = f.subs(dict(a))
    return g

def df(f, x):
    grad = []
    if type(f) == list:
        for i in range(len(f)):
            for j in range(len(x)):
                grad.append(sm.diff(f[i],x[j]))
    if type(f) !=list:
        for j in range(len(x)):
            grad.append(sm.diff(f,x[j]))
    return grad

def Jacobian(f, x):
    J = sm.zeros(len(f),len(x))
    for i, fi in enumerate(f):
        for j, s in enumerate(x):
            J[i,j] = sm.diff(fi, s)
    return J

def hessian(f, x):
    d = df(df(f,x),x)
    lol =[]
    b = 0
    for i in range(0, len(d), int(sqrt(len(d)))):
                b += int(sqrt(len(d)))
                lol.append(d[i:b])
    return lol
```

```python
def merit_function(f, s, g, h):
    k,l = 0,0
    p = sm.sympify('1/2*param_p')
    b = sm.sympify('param_b')

    for i in range(len(g)):
        k += (g[i]+s[i])**2
    for i in range(len(h)):
        l += h[i]**2

    sum_lnsi=0
    for i in range(len(g)):
        sum_lnsi += sm.sympify(f'log({s[i]})')
    merit_function = f + (k+l)*p
    return merit_function
# merit_function(f)
```

```python
def condition(s_1, x_1, s_0, x_0, p_x, p_s, param, mu, alpha, fln, s, g, h, x, param_b, param_p):

    ps = np.hstack((p_x, p_s,))
    fi_a = score_matrix(merit_function(fln, s, g, h),x+s[:len(g)]+ [param_b, param_p], np.hstack((x_1, s_1, param)) )
    fi_0 = score_matrix(merit_function(fln, s, g, h),x+s[:len(g)]+ [param_b, param_p], np.hstack((x_0, s_0, param)) )
    d_fi = np.dot(np.transpose(score_matrix(df(merit_function(fln, s, g, h),x+s[:len(g)]),x+s[:len(g)]+[param_b, param_p], np.hstack((x_0, s_0, param)))), ps)
#     print(fi_a,'>', fi_0, '+', mu*alpha*d_fi)
    return fi_a > fi_0 +mu*alpha*d_fi


def check_stopping_criteria(x_0, x, s_0, s, lamda_0, lamda, sigma_0, sigma, param, param_b, param_p, Lagrangian, h, r_opt = 1.0e-10, r_feas= 1.0e-10):
    o = np.linalg.norm(np.array(score_matrix(df(Lagrangian,x),x+s[:len(g)]+lamda[:len(h)]+sigma[:len(s)]+[param_b, param_p],np.hstack((x_0, s_0, lamda_0, sigma_0, param)))).astype(float), np.inf)
    if len(h) > 0:
        f = np.linalg.norm(np.array(score_matrix(score_matrix(h,x,x_0), s,s_0)).astype(float), np.inf)
    else: f = 1e-71
    optimality_condition = o > r_opt
    feasibility_condition = f > r_feas
    return optimality_condition or feasibility_condition
```

```python
def alphamax(s, p_s, r = 0.005):
    alpha_max = 1
    for i in range(len(s)):
        alpha = (r - 1)*s[i]/p_s[i]
        if alpha > 0:
            alpha_max = min(alpha_max, alpha)
    return alpha_max

def backtrack(x_0, s_0, p_x, p_s, alpha_max, param, fln, s,x, g, h,param_b, param_p, mu = 0.0001, ro = 0.5):
    alpha = alpha_max
    step =1.0e-10
    while condition(s_0+ np.dot(alpha, p_s), x_0+ np.dot(alpha, p_x), s_0, x_0, p_x, p_s, param, mu, alpha, fln, s, g, h, x, param_b, param_p) and step < alpha:
        alpha *= ro
    return alpha
```

```python
def update_hessian(x_1, x_0, sigma_0, lamda_0, H, x , Lagrangian , sigma, lamda):
    s_k = (np.array(x_1) - np.array(x_0)).astype(float)
    dfL = score_matrix(df(Lagrangian, x), sigma, sigma_0)
    dfL = score_matrix(dfL, lamda, lamda_0)
    y_k = (np.array(score_matrix(dfL, x, x_1)) - np.array(score_matrix(dfL, x, x_0))).astype(float)

    if np.dot(s_k.transpose() , y_k)>=0.2*np.dot( np.dot(s_k.transpose(), H) , s_k ):
        tetta = 1
    else:
        tetta = (0.8*np.dot( np.dot(s_k.transpose(), H) , s_k ))/(np.dot( np.dot(s_k.transpose(), H) , s_k )-np.dot(s_k.transpose() , y_k))
    r_k = tetta*y_k + (1-tetta)*np.dot(H, s_k)

    Hs_k = np.dot(H, s_k)
    sTHs = np.dot(s_k, Hs_k)
    term1 = np.outer(Hs_k, Hs_k) / sTHs
    term2 = np.outer(r_k, r_k) / np.dot(r_k, s_k)
    H_new = H - term1 + term2
    for i in range(len(H_new)):
        for j in range(len(H_new[i])):
            H_new[i][j] =H_new[i][j]
    return H_new
```

```python
def Hessian(Lagrangian, lamda, lamda_0, sigma, sigma_0, x, x_0, s, s_0):
    H = hessian(Lagrangian, x)
    for i in range(len(H)):
        H[i] = score_matrix(H[i], lamda, lamda_0)
        H[i] = score_matrix(H[i], sigma, sigma_0)
        H[i] = score_matrix(H[i], x, x_0)
        H[i] = score_matrix(H[i], s, s_0)
    sum = 0
    for i in range(len(H)):
        for j in range(len(H[i])):
            H[i][j] = float(H[i][j])
            if H[i][j] != 0:
                sum +=1
    if sum == 0:
        H = np.eye(len(H))
    H = np.array(H)
    return H
```

```python
def interior_point_method(x_0, lamda_0, sigma_0, param, s_0, H, param_p, param_b, Lagrangian, x, s, lamda, sigma, f, fln, ro = 0.05):
    iters=0    
    epsilon = 1.0e-5
    x_1 = np.zeros(len(x_0))
    point = [x_0]
    s_point = [score_matrix(f, x, x_0)]
    it = [iters]
    while check_stopping_criteria(x_0, x, s_0, s, lamda_0, lamda, sigma_0, sigma, param, param_b, param_p, Lagrangian, h) :
        
        Jh = np.array(score_matrix(Jacobian(h, x),x,x_0)).astype(float)
        if len(Jh) == 0:
            Jh = np.array([np.zeros(len(x_0))]).astype(float)
        Jg = np.array(score_matrix(Jacobian(g, x), x, x_0)).astype(float)
        if len(Jg) == 0:
            Jg = np.array([np.zeros(len(x_0))]).astype(float)
        I = np.eye(Jg.shape[0])

        Ssum= np.identity(len(g))
        for i in range(len(Ssum)): 
            for j in range(len(Ssum[i])):
                if i==j :
                    Ssum[i][j] = sigma_0[i]/s_0[i]

        S= np.identity(len(g))
        for i in range(len(S)): 
            for j in range(len(S[i])):
                if i==j :
                    S[i][j] = 1/s_0[i]
        e = np.ones(len(g))
        
        zeros_1 = Ssum
        zeros_2 = np.zeros((I.shape[0], Jh.shape[1])).T
        zeros_3 = np.zeros((Jh.shape[0], I.shape[1]))
        zeros_4 = np.zeros((Jh.shape[0], Jh.shape[0]))
        zeros_5 = np.zeros((Jh.shape[0], Jg.shape[0]))
        zeros_6 = np.zeros((Jg.shape[0], Jh.shape[0]))
        zeros_7 = np.zeros((Jg.shape[0], Jg.shape[0]))
        zeros_8 = np.zeros((I.shape[0], H.shape[0]))
        zeros_9 = np.zeros((I.shape[0], Jh.shape[0]))
        
        _1_row = np.hstack((H, Jh.T, Jg.T, zeros_2))
        _2_row = np.hstack((Jh, zeros_4, zeros_5, zeros_3))
        _3_row = np.hstack((Jg, zeros_6, zeros_7, I))
        _4_row = np.hstack((zeros_8, zeros_9, I, zeros_1))

        table = np.vstack((_1_row, _2_row, _3_row, _4_row))
        
        delta_Lagrange = np.array(score_matrix(score_matrix(score_matrix(df(Lagrangian,x), x, x_0), lamda, lamda_0), sigma, sigma_0)).astype(float)
        g_x = []
        for i in range(len(g)):
            g_x.append(g[i]+s[i])
        
        g_x = np.array(score_matrix(score_matrix(g_x,x,x_0), s, s_0)).astype(float)
        h_x = np.array(score_matrix(score_matrix(h,x,x_0), s, s_0)).astype(float)
        ser = sigma_0 - param[0]*np.dot(S, e)
        
        if len(h_x) ==0:
            h_x = [0]
        if len(g_x) ==0:
            g_x = [0]

        ans = np.hstack((delta_Lagrange.transpose(), h_x))
        ans = np.hstack((ans, g_x))
        ans = np.hstack((ans, ser))
        
        z = []

        for i in range(len(table)):
            if all(item == 0 for item in table[i]):
                z.append(i)
        for row in z:
            table = np.delete(table, row, axis=0)
        tab = table.transpose()
        c = []
        for i in range(len(tab)):
            if all(item == 0 for item in tab[i]):
                c.append(i)
        for row in c:
            tab = np.delete(tab, row, axis=0)
        table = tab.transpose()
        ans = np.delete(ans, z, axis = 0)

        #  solve kkt system
        solve = np.linalg.solve(table, (-ans).transpose())
        p_x, p_lamda, p_sigma, p_s = solve[:len(x_0)], solve[len(x_0):len(lamda_0)+len(x_0)], solve[len(lamda_0)+len(x_0):len(lamda_0)+len(x_0)+len(sigma_0)],solve[len(lamda_0)+len(x_0)+len(sigma_0):len(lamda_0)+len(x_0)+len(sigma_0)+len(s_0)]

        alpha_max = alphamax(s_0, p_s)
        try:
            alpha_k = backtrack(x_0, s_0, p_x, p_s, alpha_max, param, fln, s,x, g, h,param_b, param_p)
        except RuntimeError as error:
            if str(error) == 'Invalid comparison of complex numbers.':
                break
            else:
                continue
        x_1 = x_0 + np.dot(alpha_k, p_x)
        s_0 = s_0 + np.dot(alpha_k, p_s)

        alpha_sigma = alphamax(sigma_0, p_sigma)
        lamda_0 = lamda_0 + np.dot(alpha_sigma, p_lamda)
        sigma_0 = sigma_0 + np.dot(alpha_sigma, p_sigma)
     
        #  update Hessian
        H = update_hessian(x_1, x_0, sigma_0, lamda_0, H, x , Lagrangian , sigma, lamda)
        x_0= x_1
        
        param[0] = 0.5*param[0]
        
        iters = iters + 1
        point.append(x_0)
        s_point.append(score_matrix(f, x, x_0))
        it.append(iters)
        print(x_0, score_matrix(f, x, x_0))
    return iters, x_0, score_matrix(f, x, x_0), point , s_point, it
```

```python
def optimizer_IPM(f, h, g, x_0, param):
    sigma_0 = np.zeros(len(g))
    lamda_0 = np.zeros(len(h))
    s_0 = np.ones(len(g))
    x1, x2 = sm.symbols('x1 x2') 
    x = [x1, x2]
    lamda_1, lamda_2 = sm.symbols('lambda_1, lambda_2')
    lamda = [lamda_1, lamda_2]
    sigma_1, sigma_2 = sm.symbols('sigma_1, sigma_2')
    sigma = [sigma_1, sigma_2]
    s1,s2 = sm.symbols('s1 s2')
    s = [s1,s2]
    param_b, param_p = sm.symbols('param_b, param_p')
    Lagrangian, fln = Lagrangian_print(f, g, s, h, param_b, sigma, lamda)
    H = Hessian(Lagrangian, lamda, lamda_0, sigma, sigma_0, x, x_0, s, s_0)
    iters, x_0, num, point , s_point, it=interior_point_method(x_0, lamda_0, sigma_0, param, s_0, H, param_p, param_b, Lagrangian, x, s, lamda, sigma, f, fln)
    return iters, x_0, num, point , s_point, it
```

# Задача №1

```python
f = sm.sympify('x1**2')
g = sm.sympify(['1/4*x1**2+x2**2-1'])  
h = sm.sympify([])
x_0 = np.array([2,1])
param = np.array([20., 2.]) 
iters, x_0, num, point , s_point, it = optimizer_IPM(f, h, g, x_0, param)
iters, x_0
```

```python
my_df = pd.DataFrame({'iter': it, 'root': point, 'value F': s_point}).set_index('iter')
fig, ax = plt.subplots(figsize=(10, 4)) 
ax.axis('off')
tbl = table(ax, my_df, loc='center', cellLoc='center')
plt.savefig('df_as_image.png')
```
